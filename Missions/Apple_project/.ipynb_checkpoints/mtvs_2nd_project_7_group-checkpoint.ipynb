### 서론
사과는 많은 사람들에게 사랑받는 과일 중 하나이다.  
사과를 사러 마트나 시장에 가보면 어떤 사과를 골라야 맛있는 사과인지 알 수 없어 망설여질 때가 있다.  
또한 비양심적인 상인들의 포대갈이 수법으로 맛 없는 사과를 맛있는 사과로 둔갑시켜 비싼 가격에 파는 경우가 있다.  
소비자가 사과의 당도를 측정할 수 있다면 이러한 문제는 없겠지만, 사과의 당도를 측정하는 방법으로는 크게 두 가지가 있다.  
1. 착즙 당도 측정 
    * 사과의 즙을 추출하여 당도를 측정하는 방법
2. 비파괴 당도 측정
    * 근적외선을 이용한 분광분석법
    
그러나 위의 두 가지 방법은 일반 소비자가 이용하기에는 어려움이 있다.  
우리는 많은 사과의 껍질에 당도와 상관이 있는 어떤 특징이 있지 않을까 하는 궁금증을 가지고 이번 프로젝트에 임하게 되었다.  
사과의 이미지를 통해 사과의 당도를 파악할 수 있다면 우리는 스마트폰 카메라로 맛있는 사과를 구매할 수 있을것이다.

# 데이터 분석
데이터 분석에 앞서 분류된 데이터 중 문제점을 발견하였다.  
당도 등급이 B인 사과임에도 불구하고 A나 C가 분류된 디렉터리에 데이터가 들어가있어서  
다시 분류할 필요가 있었다.

```python
import pandas as pd
import numpy as np
from pathlib import Path
from glob import glob
import shutil
import os
from tqdm import tqdm
import matplotlib.pyplot as plt
import json
import seaborn as sns
plt.rcParams['font.family'] = 'NanumGothic'
```

```python
# 사용될 path들을 지정
root = Path("D:/").resolve()
base = root / "Apple_data"
data = base / "dataset_mini"
label = base / "labels_mini"
mod_data = base / 'dataset_min_cate'
cate_data = base / "dataset_cate"
grade_data = base / "dataset_grade"
```

```python
apple_kind = {'SG':'시나노골드', 'AR':'아리수', 'HR':'홍로', 'HJ': '후지'}
columns = ['파일이름','파일경로','당도_등급',
           '품종','착즙당도','비파괴당도',
           '토양_전기전도도','토양_온도','토양_습도',
           '토양_수분장력','온도','습도','일사량']
```

```python
# 필요 함수

# 메타데이터 수집 함수
def get_meta (j_data, dir_name, file_name):
    tmp = [file_name, dir_name]
    tmp.append(j_data['annotations']['sugar_grade'])
    tmp.append(j_data['collection']['apple_kind'])
    tmp.append(j_data['collection']['sugar_content'])
    tmp.append(j_data['collection']['sugar_content_nir'])
    tmp.append(j_data['collection']['soil_ec'])
    tmp.append(j_data['collection']['soil_temper'])
    tmp.append(j_data['collection']['soil_humidty'])
    tmp.append(j_data['collection']['soil_potential'])
    tmp.append(j_data['collection']['temperature'])
    tmp.append(j_data['collection']['humidity'])
    tmp.append(j_data['collection']['sunshine'])
    return tmp

# 디렉터리 생성 함수
def mkdir(fname, path):
    if os.path.exists(path / fname):
        print(f'이미 {fname}디렉터리가 존재합니다.')
    else:
        os.mkdir(path / fname)
        print(f'{fname}디렉터리를 생성하였습니다.')
```

```python
# 데이터셋의 메타데이터를 불러오기
result = [columns]
for fold in glob(str(label / '*')):
    dir_name = fold.split('\\')[-1]
    for file in tqdm(glob(str(fold + '\*'))):
        file_name = file.split('\\')[-1]
        with open(file, 'r', encoding='utf-8') as f:
            file_data = json.load(f)
            result.append(get_meta(file_data, dir_name, file_name))
```

```python
# 메타데이터를 데이터프레임화
df = pd.DataFrame(result[1:], columns=result[0])
df
```

```python
# 각 등급에 맞게 데이터가 분배되어있지 않음을 확인
df.loc[df['당도_등급'] =='B']['파일경로'].unique()
```

```python
# 파일 위치를 등급에 맞게 재편성
for grade in ['A', 'B', 'C']:
    print('=' * 10 + grade + '='*10)
    for file, path, grade in df.loc[df['당도_등급'] == grade][['파일이름', '파일경로','당도_등급']].values:
        if grade != path[-1]:
            # ex) 시나노골드A -> 시나노골드B
            shutil.move(label / path / str(file), label / str(path[:-1] + grade))
```

```python
# 데이터셋의 메타데이터 저장
df.to_csv(base / 'apple.csv')
```

# 학습 시키기

```python
import os
from PIL import Image
import tensorflow as tf
import tensorflow_datasets as tfds
from tensorflow import keras
from tensorflow.keras.applications.mobilenet_v2 import MobileNetV2
from tensorflow.keras.applications.vgg16 import VGG16
from tensorflow.keras.applications.resnet_v2 import ResNet50V2
from tensorflow.keras.preprocessing.image import img_to_array
from tensorflow.keras.applications.efficientnet_v2 import EfficientNetV2S
from tensorflow.keras.preprocessing.image import load_img
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense
```

```python
cate_dir = cate_data
data_dir = mod_data
```

```python
# 전체 이미지 데이터 수 확인
image_count = len(list(data_dir.glob('*/*.jpg')))
print(image_count)
```

# 데이터 증강

```python
IMG_SIZE = 220

resize_and_rescale = tf.keras.Sequential([
  keras.layers.experimental.preprocessing.Resizing(IMG_SIZE, IMG_SIZE),
  keras.layers.experimental.preprocessing.Rescaling(1./255)
])
```

```python
data_augmentation = Sequential([
  keras.layers.experimental.preprocessing.RandomFlip("horizontal_and_vertical"),
  keras.layers.experimental.preprocessing.RandomRotation(0.2),
])
```

```python
def random_invert_img(x, p=0.5):
    if  tf.random.uniform([]) < p:
        x = (1-x)
    else:
        x
    return x
```

```python
def random_invert(factor=0.5):
    return keras.layers.Lambda(lambda x: random_invert_img(x, factor))

random_invert = random_invert()
```

```python
def data_preprop(x):
    aug_ds = x.map(lambda x, y: (resize_and_rescale(x), y))
#     aug_ds = aug_ds.map(lambda x, y: (random_invert_img(x), y))
    aug_ds = aug_ds.map(lambda x, y: (data_augmentation(x), y))
#     aug_ds = aug_ds.map(lambda x, y: (tf.image.rgb_to_grayscale(x), y))
    return aug_ds
```

```python
# Add the image to a batch
image = np.array(Image.open('./hj.jpg'))
image = resize_and_rescale(image)
image = tf.expand_dims(image, 0)
```

```python
plt.figure(figsize=(10, 10))
for i in range(9):
#     augmented_image = random_invert(image)
    augmented_image = data_augmentation(image)
    ax = plt.subplot(3, 3, i + 1)
    plt.imshow(augmented_image[0])
    plt.axis("off")
```

```python
def visualize(original, augmented):
  fig = plt.figure()
  plt.subplot(1,2,1)
  plt.title('Original image')
  plt.imshow(original)

  plt.subplot(1,2,2)
  plt.title('Augmented image')
  plt.imshow(augmented)
```

```python
grayscaled = tf.image.rgb_to_grayscale(image)
visualize(image[0], tf.squeeze(grayscaled))
_ = plt.colorbar()
```

```python
batch_size = 32
img_height = 220
img_width = 220

train_ds = tf.keras.preprocessing.image_dataset_from_directory(
  cate_dir,
  validation_split=0.2,
  subset="training",
  seed=123,
  image_size=(img_height, img_width),
  batch_size=batch_size)

val_ds = tf.keras.preprocessing.image_dataset_from_directory(
  cate_dir,
  validation_split=0.2,
  subset="validation",
  seed=123,
  image_size=(img_height, img_width),
  batch_size=batch_size)

class_names = train_ds.class_names
print(class_names)
```

```python
AUTOTUNE = tf.data.AUTOTUNE

train_ds = train_ds.cache().prefetch(buffer_size=AUTOTUNE)
val_ds = val_ds.cache().prefetch(buffer_size=AUTOTUNE)
```

```python
aug_ds = data_preprop(train_ds)
```

```python
aug_ds = train_ds.map(lambda x, y: (random_invert_img(x), y))
aug_ds = aug_ds.map(lambda x, y: (data_augmentation(x), y))
aug_ds = aug_ds.map(lambda x, y: (tf.image.rgb_to_grayscale(x), y))
```

```python
plt.figure(figsize=(10, 10))
for images, labels in aug_ds.take(1):
    print(images[0])
    for i in range(9):
        ax = plt.subplot(3, 3, i + 1)
        plt.imshow(images[i])
        plt.title(class_names[labels[i]])
        plt.axis("off")
```

```python
# 이미지 처리를 위한 MobileNetV2모델
mob_base = MobileNetV2(include_top=False, input_shape=(220, 220, 3))
vgg_base = VGG16(include_top=False, input_shape=(220, 220, 3))
res_base = ResNet50V2(include_top=False, input_shape=(220, 220, 3))
```

```python
vgg_base.summary()
```

```python
# 출력층 모델 생성
model = Sequential()
# model.add(mob_base)
model.add(vgg_base)
# model.add(res_base)
model.add(Flatten())
model.add(Dense(256, activation='relu'))
model.add(Dense(3, activation='softmax'))
model.summary()
```

```python
model.compile(loss=tf.keras.losses.SparseCategoricalCrossentropy(),
              optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),
              metrics=['accuracy'])
```

```python
# 학습 중단하기
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.callbacks import EarlyStopping

early_stopping_callback = EarlyStopping(monitor='val_loss', patience=20)

modelpath = "./save/apple_project/cate_vgg16_{epoch:02d}-{val_accuracy:.4f}.hdf5"

checkpointer = ModelCheckpoint(filepath=modelpath, monitor='val_loss', verbose = 1)
```

```python
history = model.fit(train_ds, epochs = 50, validation_data=val_ds, callbacks=[checkpointer])
```

```python
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

loss=history.history['loss']
val_loss=history.history['val_loss']

epochs_range = range(50)

plt.figure(figsize=(8, 8))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.savefig('./cate_vgg16_2.jpg')
plt.show()
```
