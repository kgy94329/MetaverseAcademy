```python
import cv2
import sys
import os
import matplotlib.pyplot as plt
```

# cv2 이미지 불러오기
openCV는 이미지 파일을 불러오면 기본적으로 BGR형식으로 파일을 생성한다.  
그러므로, RGB형태로 파일을 처리할 때는 반드시 형식을 변경해줄 필요가 있다.

```python
img = cv2.imread('img/images.jpg')
img
```

```python
# 불러온 img파일을 그대로 출력
plt.imshow(img)
plt.show()
```

```python
# img파일을 RGB이미지로 변경
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

```python
# img파일을 grayscale하기
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img, cmap='gray')
plt.show()
```

```python
# 편집한 이미지 저장하기
cv2.imwrite('img/images_gray.jpg', img)
```

```python
# 기본 이미지와 gray이미지 크기 확인하기
img.shape, img_gray.shape
```

# Video파일 불러오기

```python
cap = cv2.VideoCapture('img/city.mp4')

while True:
    ret, frame = cap.read()

    # 불러온 영상이 없으면 처리 종료
    if not ret:
        break
    
    # 매 프레임마다 영상을 재생
    cv2.imshow('frame', frame)

    # esc가 입력되면 처리를 종료한다.
    # 기본 1초에 30프레임이 재생되므로, waitKey값을 30으로 입력하면
    # 영상이 원래 속도로 재생된다.
    if cv2.waitKey(30) == 27:
        break

# 영상 처리 종료
cap.release()
# 창 닫기
cv2.destroyAllWindows()
```
