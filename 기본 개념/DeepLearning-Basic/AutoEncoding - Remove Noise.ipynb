# 데이터 전처리
- Min-Max normalization

```python
# Data load
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()

# Min-Max normalization & change data type (int to float)
x_train = x_train.astype('float32') / 255.
x_test = x_test.astype('float32') / 255.

print (x_train.shape)
print (x_test.shape)
print (y_train.shape)
print (y_test.shape)
```

```python
# Min-Max 정규화로 각 값의 최대값과 최소값이 1과 0임을 확인
x_train.max(), x_train.min(), x_test.max(), x_test.min()
```

```python
# 흑백이미지를 컬러 이미지로 변경 (28, 28) -> (28, 28, 3)
from skimage import color
x_train = color.gray2rgb(x_train)
x_test = color.gray2rgb(x_test)
x_train.shape, x_test.shape
```

```python
# 이미지 확인
plt.imshow(x_train[0])
```

# 노이즈 추가
(28, 28)크기의 배열에 0~1 사이값을 무작위로 생성하고자 함.  
* 가우스 분포에 따른 난수를 생성, 0.5는 평균값, 3.0은 표준 편차값을 의미  
  246522는 이미지 값이 갖는 모든 요소의 수를 의미 (362 * 227 * 3)
* noise의 shape는 (246522, ) 형태로, 한 행으로 이루어져있는 상태이므로, array의 형태가 아니라서 에러가 발생

```python
noise = np.random.normal(0.5, 3.0, (28,28,3))
print(noise.shape)
noise
```

```python
# 노이즈 이미지를 원본데이터와 결합
# 노이즈 데이터의 값이 강해서 원본데이터가 보이지 않는 것을 알 수 있다.

plt.subplot(1, 3, 1)
plt.imshow(noise)
plt.title('Noise')

plt.subplot(1, 3, 2)
plt.imshow(x_train[0])
plt.title('TrainData')

plt.subplot(1, 3, 3)
plt.imshow(x_train[0] + noise)
plt.title('Noise + Train')
```

```python
# noise값을 min-max정규화
norm_noise = (noise - noise.min()) / (noise.max() - noise.min())
print(norm_noise.max(), norm_noise.min())
norm_noise
```

```python
# 정규화된 노이즈와 원본데이터를 결합
# min-max정규화된 노이즈 색이 옅어진것을 알 수 있다
plt.subplot(1, 3, 1)
plt.imshow(norm_noise)
plt.title('Noise')

plt.subplot(1, 3, 2)
plt.imshow(x_train[0])
plt.title('TrainData')

plt.subplot(1, 3, 3)
plt.imshow(x_train[0] + norm_noise)
plt.title('Noise + Train')
```

# 모델 구축

```python
# 직접 구축한 모델
from tensorflow.keras.layers import Input, Conv2D, Flatten, Dense, Reshape, Conv2DTranspose
from tensorflow.keras.models import Model

inputs = Input(shape=(28,28,3))
x = Conv2D(32, (3, 3), activation = 'relu', padding='same')(inputs) # (28 - 3) + 2 +1 
x = Conv2D(64, (3, 3), activation = 'relu', padding='same')(x) # 28, 28, 64
x = Flatten()(x) # 28 * 28 * 64
last = Dense(10)(x) # 10개 * ((28*28*64 = 50176) + 바이어스1) -> 501770

x = Dense(28*28*64)(last)
x = Reshape((28, 28, 64))(x)
x = Conv2DTranspose(64, (3, 3), activation = 'relu', padding='same')(x) # (28 - 3) + 2 +1 
x = Conv2DTranspose(32, (3, 3), activation = 'relu', padding='same')(x) # 28, 28, 64
x = Conv2DTranspose(3, (1, 1), activation = 'sigmoid', padding='same')(x)

#모델생성
model = Model(inputs, x)
model.summary()
```

```python
model.compile(loss = 'mae', optimizer='adam', metrics=['accuracy'])
```

```python
# 노이즈를 입힌 데이터를 독립변수로, 원본 데이터를 종속변수로 둔다
hist = model.fit(col_x_train_noise, x_train, validation_data=(col_x_test_noise, x_test),
                epochs = 10)
```

```python
result = model.predict(col_x_test_noise)
```

# 규형이형 모델

```python
class Denoise(Model):
  def __init__(self, in_ch = 3):
    super(Denoise, self).__init__()
    self.encoder = tf.keras.Sequential([
      layers.Input(shape=(28, 28, in_ch)), 
      layers.Conv2D(32, 3, activation='relu', padding='same', strides=1),
      layers.Conv2D(16, (3,3), activation='relu', padding='same', strides=2),
      layers.Conv2D(8, (3,3), activation='relu', padding='same', strides=2)])

    self.decoder = tf.keras.Sequential([
        # Conv2DTranspose = 기존 행렬을 전치하는 레이어
      layers.Conv2DTranspose(8, kernel_size=3, strides=2, activation='relu', padding='same'),
      layers.Conv2DTranspose(16, kernel_size=3, strides=2, activation='relu', padding='same'),
      layers.Conv2DTranspose(32, kernel_size=3, strides=1, activation='relu', padding='same'),
      layers.Conv2D(in_ch, kernel_size=1, activation='sigmoid', padding='same')])

  def call(self, x):
    encoded = self.encoder(x)
    # print(encoded.shape)
    decoded = self.decoder(encoded)
    # print(decoded.shape)
    return decoded
```

```python
autoencoder = Denoise()
autoencoder.compile(optimizer='adam', loss=losses.MeanSquaredError(),
                   metrics=['accuracy'])

autoencoder.fit(col_x_train_noise, x_train,
                epochs=20,
                shuffle=True,
                validation_data=(col_x_test_noise, x_test))
```

```python
result = autoencoder.predict(col_x_test_noise)
```

```python
# Tensorflow 가이드 모델
latent_dim = 64 

class Autoencoder(Model):
  def __init__(self, encoding_dim):
    super(Autoencoder, self).__init__()
    self.latent_dim = latent_dim   
    # 인코딩 레이어
    self.encoder = tf.keras.Sequential([
      layers.Flatten(),
      layers.Dense(latent_dim, activation='relu'),
    ])
    # 디코딩 레이어
    self.decoder = tf.keras.Sequential([
      layers.Dense(784, activation='sigmoid'),
      layers.Reshape((28, 28))
    ])

  def call(self, x):
    encoded = self.encoder(x)
    decoded = self.decoder(encoded)
    return decoded

autoencoder = Autoencoder(latent_dim)
```

```python
autoencoder.compile(optimizer='adam', loss=losses.MeanSquaredError())
```

```python
autoencoder.fit(x_train, x_train,
                epochs=10,
                shuffle=True,
                validation_data=(x_test, x_test))
```

```python
encoded_imgs = autoencoder.encoder(x_test).numpy()
decoded_imgs = autoencoder.decoder(encoded_imgs).numpy()
```

```python
n = 10
plt.figure(figsize=(20, 4))
for i in range(n):
  # display original
  ax = plt.subplot(2, n, i + 1)
  plt.imshow(x_test[i])
  plt.title("original")
  plt.gray()
  ax.get_xaxis().set_visible(False)
  ax.get_yaxis().set_visible(False)

  # display reconstruction
  ax = plt.subplot(2, n, i + 1 + n)
  plt.imshow(decoded_imgs[i])
  plt.title("reconstructed")
  plt.gray()
  ax.get_xaxis().set_visible(False)
  ax.get_yaxis().set_visible(False)
plt.show()
```
