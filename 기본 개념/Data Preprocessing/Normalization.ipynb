# 데이터 정규화(Normalization)

머신러닝에서 사용되는 데이터는 그 값이 너무 크거나 작지 않고 적당한 범위 (-1 ~ 1)에 있어야 모델의 정확도가 높아진다고 알려져있다.  
그래서 학습 전에 미리 값이 적당한 범위를 유지하도록 모델에 입력할 데이터를 변환하는 것이 좋다.  
우리는 이 변환 작업을 데이터 정규화라고 한다.  

* 데이터 정규화
    * 정규화(normalization) - Max-Min normalization  
        입력 변수 x를 최소값을 0, 최대값을 1이 되도록 변환한다.  
        $x_{new} =$ $x - x_{min} \over x_{max} - x_{min}$
        
    * 표준화(Standardization)  
        입력 변수 x를 평균이 0, 분산이 1인 정규 분포(가우시안 분포)가 되도록 변환한다.  
        $z = $ $X - \mu \over \sigma $  
        $z = $ 입력된 데이터  
        $ \mu = $ 평균  
        $ \sigma = $ 표준편차  
        이 작업을 Z-Score Normalization이라고도 한다.  

* 주의사항
    * 이 기법들은 어떤 알고리즘을 이용하느냐에 따라 다르다.  

        * 효과가 좋은 알고리즘  
         1. 선형회귀
         2. 로지스틱 회귀
         3. 서포트 벡터 머신
        * 효과가 좋지 않은 알고리즘  
         1. 결정 트리 알고리즘

    * 두 가지 기법 중 정규화는 이상값의 영향을 쉽게 받는다.  
        따라서, 데이터에 이상값이 포함되었을 가능성이 높다면 표준화를 적용한다.
    * 이미지 데이터와 같이 최솟값과 최대값을 미리 알 수 있는 경우에는 정규화를 사용하는 경우가 많다.

# 파이썬의 데이터 정규화 방법
### 1. Max-Min normalization

```python
import numpy as np

# 랜덤 함수를 통해 무작위로 데이터 추출
dummy = (np.random.random(100)*100).astype(np.int32)
print('생성된 난수', dummy, sep = '\n')

print()
# 정규화 적용
norm_dummy = (dummy - np.min(dummy)) / (np.max(dummy) - np.min(dummy))
print('정규화된 난수', norm_dummy, sep='\n')
```

### 2. Z-Score Normalization

```python
# 랜덤 함수를 통해 무작위로 데이터 추출
dummy = (np.random.random(100)*100).astype(np.int32)
print('생성된 난수', dummy, sep = '\n')
print('정규화된 난수', z_dummy, sep='\n')
print()
print('난수의 평균 = {} \n난수의 표준편차 = {}'.format(np.mean(dummy), np.std(dummy)))
print()
# 정규화 적용
z_dummy = (dummy - np.mean(dummy)) / np.std(dummy)
print('정규화된 난수', z_dummy, sep='\n')
print()
print('정규화된 난수의 평균 = {} \n정규화된 난수의 표준편차 = {}'.format(np.mean(z_dummy), np.std(z_dummy)))
```

### 3. 이미지 데이터의 정규화
이미지 데이터를 정규화해도 이미지에는 아무 변화가 없음을 알 수 있다.

```python
from PIL import Image
import matplotlib.pyplot as plt

plt.figure(figsize = (10, 10))

# 원본데이터
img = np.array(Image.open('./data/Lenna.png').convert('RGB'))
print('원본데이터 요소 확인', img, sep='\n')
plt.subplot(2, 1, 1)
plt.imshow(img)
plt.title('Original Data')
print()
# 이미지의 정규화
norm_img = img.astype(np.float32) / 255
print('정규화 데이터 요소 확인', norm_img, sep='\n')
plt.subplot(2, 1, 2)
plt.imshow(norm_img)
plt.title('Normalized Data')
```
